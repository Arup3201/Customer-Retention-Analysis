*Findings*
1. Most customers are using the ABC bank branch of `France`. `Spain` and `Germany` have almost same amount of customers.
2. Although there is not much difference in `Male` and `Female` customer counts in the ABC bank, still there are more `Male` customers than `Female`.
3. Products with number `1` and `2` are the most purchased product. Other 2 products are rarely purchased.
4. More than 6000 customers are using credit card. And more than 2000 customers are not using credit card.
5. Active and non-active customer count is very close.

```python
sizes = [df.churn[df['churn']==1].count(), df.churn[df['churn']==0].count()]
labels = ['Churned', 'Not Churned']
figure, axes = plt.subplots()
axes.pie(sizes, labels=labels, autopct = '%1.2f%%', colors=palette[-2:])
plt.legend(loc='upper right')
plt.title("Churned VS Not Churned", size = 12)
plt.show()
```

*Findings*
- More than 20% customers are churning.

### Feature effect on churn rate

```python
df.columns
```

#### Continuous features

- credit_score
- age
- tenure
- balance
- estimated_salary

```python
sns.histplot(x=df['credit_score'], hue=df['churn'], kde=True, palette=palette[-2:])
plt.show()
```

```python
sns.histplot(x=df['age'], hue=df['churn'], kde=True, palette=palette[-2:])
plt.show()
```

```python
sns.histplot(x=df['balance'], hue=df['churn'], kde=True, palette=palette[-2:])
plt.show()
```

```python
sns.histplot(x=df['estimated_salary'], hue=df['churn'], kde=True, palette=palette[-2:])
plt.show()
```

*Findings*
1. `age` have a clear impact on the churn rate of users. Other continuous features do not contribute much on the churn rate of users. Because the churn rate is exactly proportionate with the volume of customer in that bar.
2. Users with more `age` are more likely to churn than young users.

*Finding*
1. Highest churn rate is at `Germany` and `France` has lowest churn rate.
2. `Female`s are more likely to churn than `Male`s.
3. `tenure` has very less impact on churn rate.
4. Product 3 and 4 are more likely to churn in fact most customers who buy these products churn. Product 2 has very less churn rate.
5. Active members are less likely to churn.
